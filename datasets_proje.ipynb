```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('population_by_country_2020.csv')
data['Density (P/Km²)'] = pd.to_numeric(data['Density (P/Km²)'], errors='coerce')
top_10_density = data.nlargest(10, 'Density (P/Km²)')

plt.barh(top_10_density['Country (or dependency)'], top_10_density['Density (P/Km²)'], color='skyblue')
plt.xlabel('Density (P/Km²)')
plt.title('Top 10 Most Densely Populated Countries (2020)')
plt.gca().invert_yaxis()
plt.show()
```

```python
data['Population (2020)'] = pd.to_numeric(data['Population (2020)'], errors='coerce')
top_10_population = data.nlargest(10, 'Population (2020)')

# Grafik çizimi
plt.figure(figsize=(12, 8))
plt.barh(top_10_population['Country (or dependency)'], top_10_population['Population (2020)'], color='coral')
plt.xlabel('Population (2020)')
plt.title('Top 10 Most Populated Countries (2020)')
plt.gca().invert_yaxis()
plt.show()
```
