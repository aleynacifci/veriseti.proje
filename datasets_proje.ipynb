```python
import pandas as pd

df = pd.read_csv('/content/top50.csv', encoding="latin1") #bilinmeyen karakterler için latin dedik

print(df.to_string())
```

```python
import matplotlib.pyplot as plt

artist_counts = df['Artist.Name'].value_counts().head(50)

plt.figure(figsize=(10,6))
plt.bar(artist_counts.index, artist_counts.values, color='brown')
plt.xlabel('Artist Name')
plt.ylabel('Number of Songs in Top 50')
plt.title('Top 50 Artist by Number of Songs')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Genre'], color='green')
plt.xlabel('Genre')
plt.ylabel('Number of songs')
plt.title('Song Genres')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
popularity = df['Popularity'].value_counts().head(50)
df = df.sort_values(by='Popularity', ascending=False)
plt.figure(figsize=(10,6))
plt.bar( df['Track.Name'],  df['Popularity'].values, color='navy')
plt.xlabel('Popularity')
plt.ylabel('Number of Songs Popularity')
plt.title('Number of Songs Popularity')
plt.xticks(rotation=90)
plt.show()
```
